```python
# Importing necessary libraries
import pandas as pd
import numpy as np

# Load the data from CSV file
car = pd.read_csv('quikr_car.csv')

# Backup the original data before cleaning
backup = car.copy()

# Converting 'year' column to numeric and cleaning 'Price' and 'kms_driven'
car['year'] = car['year'].str.isnumeric()  # Keep rows where 'year' is numeric
car = car[car['year']]  # Remove invalid 'year' entries
car['year'] = car['year'].astype(int)

# Cleaning the 'Price' column
car = car[car['Price'] != 'Ask For Price']  # Remove 'Ask For Price'
car['Price'] = car['Price'].str.replace(',', '').astype(int)

# Cleaning 'kms_driven'
car['kms_driven'] = car['kms_driven'].str.split().str.get(0).str.replace(',', '')
car = car[car['kms_driven'].str.isnumeric()]
car['kms_driven'] = car['kms_driven'].astype(int)

# Removing rows with missing fuel_type
car = car[~car['fuel_type'].isna()]

# Removing rows where 'Price' is above a certain threshold
car = car[car['Price'] < 6000000]

# Extracting car company from the 'name' column
car['company'] = car['name'].str.split().str.get(0)

# Reset the index after cleaning
car = car.reset_index(drop=True)

# Saving cleaned data to a new CSV file
car.to_csv('Cleaned_Car_data.csv')
```

```python
# Importing necessary visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Basic visualizations

# Plotting the frequency of car companies
car['company'].value_counts().plot(kind='bar')
plt.title("Car Companies Distribution")
plt.xlabel("Company")
plt.ylabel("Count")
plt.show()

# Plotting the average price per company
car.groupby('company')['Price'].mean().plot(kind='bar')
plt.title("Average Price per Company")
plt.xlabel("Company")
plt.ylabel("Average Price")
plt.show()

# Plotting the distribution of car prices
car['Price'].plot(kind='hist', bins=50)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.show()

# Boxplot of Price by Company
plt.figure(figsize=(15,7))
sns.boxplot(x='company', y='Price', data=car)
plt.xticks(rotation=40, ha='right')
plt.title("Price Distribution by Company")
plt.show()

# Swarmplot for Price vs Year
plt.figure(figsize=(20,10))
sns.swarmplot(x='year', y='Price', data=car)
plt.xticks(rotation=40, ha='right')
plt.title("Price vs Year")
plt.show()
```
